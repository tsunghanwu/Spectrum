```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Dataset Overview

```python
df = pd.read_csv("sample.csv", skiprows=2)
df = df.iloc[:,1:3]
df.columns = ["Time", "Intensity"]
```

```python
df.head()
```

```python
df.info()
```

## Original Reflectometry Signal

```python
plt.plot(df["Time"], df["Intensity"])
```

## Deploy Rolling Mean on Reflectometry Signal

```python
rm20 = df["Intensity"].rolling(window=20).mean()
rm40 = df["Intensity"].rolling(window=40).mean()
rm60 = df["Intensity"].rolling(window=60).mean()
rm80 = df["Intensity"].rolling(window=80).mean()
```

```python
plt.plot(df["Time"], df["Intensity"])
plt.plot(df["Time"][19:], rm20[19:]+10000, label="Window = 20")
plt.plot(df["Time"][39:], rm40[39:]+20000, label="Window = 40")
plt.plot(df["Time"][59:], rm60[59:]+30000, label="Window = 60")
plt.plot(df["Time"][79:], rm80[79:]+40000, label="Window = 80")
plt.legend()
```

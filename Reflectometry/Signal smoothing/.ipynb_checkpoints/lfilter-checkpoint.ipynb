```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib inline
```

## Dataset Overview

```python
df = pd.read_csv("sample.csv", skiprows=2)
df = df.iloc[:,1:3]
df.columns = ["Time", "Intensity"]
```

```python
df.head()
```

## Original Reflectometry Signal

```python
plt.plot(df["Time"], df["Intensity"])
```

## lfilter Deployment

```python
b, a = signal.butter(1, 0.1)
zi = signal.lfilter_zi(b, a)
f = signal.lfilter(b, a, df["Intensity"], zi=zi*df["Intensity"].iloc[0])

zi*df["Intensity"].iloc[0]
```

```python
f
```

```python
plt.plot(df["Time"], f[0])
```

## Compare Different Filter Freq (1st order)

```python
for i in range(1,5):
    b, a = signal.butter(1, 10**(-i))
    zi = signal.lfilter_zi(b, a)
    f = signal.lfilter(b, a, df["Intensity"], zi=zi*df["Intensity"].iloc[0])
    plt.plot(df["Time"], f[0], label="Freq = {}".format(10**(-i)))

plt.legend()
```
